# Data Standardization
The process of standardizing tthe data to a common format and common range

```python
import sklearn.datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(dataset)
```

```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
df.head()
```

```python
df.shape
```

```python
X = df
Y = dataset.target
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Standardize the data

```python
print(dataset.data.std())
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_standardized = scaler.transform(X_train)
```

```python
print(X_train_standardized)
```

```python
X_test_standardized = scaler.transform(X_test)
```

```python
print(X_train_standardized.std())
```

```python
print(X_test_standardized.std())
```

# Label Encoding
- Converting the labels into numeric form

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# loading the data from csv file to pandas dataFrame
cancer_data = pd.read_csv('/content/data.csv')
```

```python
# first 5 rows of the dataframe
cancer_data.head()
```

```python
# finding the count of different labels
cancer_data['diagnosis'].value_counts()
```

```python
# load the Label Encoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
# appending the labels to the DataFrame
cancer_data['target'] = labels
```

```python
cancer_data.head()
```

0 --> Benign

1 --> Malignant

```python
cancer_data['target'].value_counts()
```

# Imbalance Dataset

```python
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_data
```

```python
credit_card_data['Class'].value_counts()
```

```python
# separating the legit and fraudulent transactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit_sample = legit.sample(n=3)
```

```python
print(legit_sample.shape)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset
```

```python
new_dataset['Class'].value_counts()
```
